## 2. Contexto

No exemplo fornecido, o objeto detailLines contém um menuItem. Ele também pode
conter instâncias de:

● discount

● serviceCharge

● tenderMedia

● errorCode

Transcreva o JSON para tabelas SQL. A implementação deve fazer sentido para
operações de restaurante.

#### Esse é o diagrama que representa a implementação SQL abaixo

```python
CREATE TABLE guestChecks (
    guestCheckId BIGINT PRIMARY KEY,
    chkNum INT,
    opnBusDt DATE,
    opnUTC DATETIME,
    opnLcl DATETIME,
    clsdBusDt DATE,
    clsdUTC DATETIME,
    clsdLcl DATETIME,
    lastTransUTC DATETIME,
    lastTransLcl DATETIME,
    lastUpdatedUTC DATETIME,
    lastUpdatedLcl DATETIME,
    clsdFlag BOOLEAN,
    gstCnt INT,
    subTtl DECIMAL(10, 2),
    nonTxblSlsTtl DECIMAL(10, 2),
    chkTtl DECIMAL(10, 2),
    dscTtl DECIMAL(10, 2),
    payTtl DECIMAL(10, 2),
    balDueTtl DECIMAL(10, 2),
    rvcNum INT,
    otNum INT,
    ocNum INT,
    tblNum INT,
    tblName VARCHAR(50),
    empNum INT,
    numSrvcRd INT,
    numChkPrntd INT
);

CREATE TABLE taxes (
    taxId BIGINT AUTO_INCREMENT PRIMARY KEY,
    guestCheckId BIGINT,
    taxNum INT,
    txblSlsTtl DECIMAL(10, 2),
    taxCollTtl DECIMAL(10, 2),
    taxRate DECIMAL(5, 2),
    type INT,
    FOREIGN KEY (guestCheckId) REFERENCES guestChecks(guestCheckId)
);


CREATE TABLE detailLines (
    guestCheckLineItemId BIGINT PRIMARY KEY,
    guestCheckId BIGINT,
    rvcNum INT,
    dtlOtNum INT,
    dtlOcNum INT,
    lineNum INT,
    dtlId INT,
    detailUTC DATETIME,
    detailLcl DATETIME,
    lastUpdateUTC DATETIME,
    lastUpdateLcl DATETIME,
    busDt DATE,
    wsNum INT,
    dspTtl DECIMAL(10, 2),
    dspQty DECIMAL(10, 2),
    aggTtl DECIMAL(10, 2),
    aggQty DECIMAL(10, 2),
    chkEmpId INT,
    chkEmpNum INT,
    svcRndNum INT,
    seatNum INT,
    FOREIGN KEY (guestCheckId) REFERENCES guestChecks(guestCheckId)
);


CREATE TABLE menuItem (
    guestCheckLineItemId BIGINT PRIMARY KEY,
    miNum INT,
    modFlag BOOLEAN,
    inclTax DECIMAL(10, 2),
    activeTaxes VARCHAR(50),
    prcLvl INT,
    FOREIGN KEY (guestCheckLineItemId) REFERENCES detailLines(guestCheckLineItemId)
);


CREATE TABLE discount (
    guestCheckLineItemId BIGINT PRIMARY KEY,
    discountId INT,
    discountAmt DECIMAL(10, 2),
    discountType VARCHAR(50),
    FOREIGN KEY (guestCheckLineItemId) REFERENCES detailLines(guestCheckLineItemId)
);


CREATE TABLE serviceCharge (
    guestCheckLineItemId BIGINT PRIMARY KEY,
    serviceChargeId INT,
    serviceAmt DECIMAL(10, 2),
    serviceType VARCHAR(50),
    FOREIGN KEY (guestCheckLineItemId) REFERENCES detailLines(guestCheckLineItemId)
);

CREATE TABLE tenderMedia (
    guestCheckLineItemId BIGINT PRIMARY KEY,
    tenderMediaId INT,
    tenderAmt DECIMAL(10, 2),
    tenderType VARCHAR(50),
    FOREIGN KEY (guestCheckLineItemId) REFERENCES detailLines(guestCheckLineItemId)
);

CREATE TABLE errorCode (
    guestCheckLineItemId BIGINT PRIMARY KEY,
    errorCodeId INT,
    errorMsg VARCHAR(255),
    FOREIGN KEY (guestCheckLineItemId) REFERENCES detailLines(guestCheckLineItemId)
);
```

### Explicação de cada relacionamento aplicado:

##### GuestCheck ↔ Tax
Tipo: 1:N
Descrição: Um pedido pode ter vários impostos aplicados (ex.: impostos estaduais e municipais).

##### GuestCheck ↔ DetailLine
Tipo: 1:N
Descrição: Um pedido pode conter várias linhas detalhadas, representando itens, descontos ou taxas.

##### DetailLine ↔ MenuItem
Tipo: 1:1
Descrição: Cada linha pode conter um item específico do menu (ex.: hambúrguer ou refrigerante).

##### DetailLine ↔ Discount
Tipo: 1:1
Descrição: Cada linha pode ter um desconto aplicado (ex.: promoção de 10%).

##### DetailLine ↔ ServiceCharge
Tipo: 1:1
Descrição: Cada linha pode incluir uma taxa de serviço (ex.: taxa de 10% para entrega).

##### DetailLine ↔ TenderMedia
Tipo: 1:1
Descrição: Cada linha pode ser paga com um método específico (ex.: cartão ou dinheiro).

##### DetailLine ↔ ErrorCode
Tipo: 1:1
Descrição: Cada linha pode conter um erro registrado (ex.: erro ao calcular imposto).

## 3. Descreva a abordagem escolhida em detalhes. Justifique a escolha.

A abordagem escolhida organiza os dados do JSON original em uma estrutura relacional que reflete a hierarquia natural do modelo e as necessidades operacionais de um restaurante. O GuestCheck atua como a entidade central, representando o pedido completo, enquanto detailLines captura os componentes detalhados, como itens do menu, descontos, taxas de serviço e métodos de pagamento. Esses detalhes foram distribuídos em tabelas especializadas para evitar redundância, facilitar a manutenção e melhorar a escalabilidade.

Optamos por normalizar os dados, criando relações claras entre GuestCheck e entidades como Tax, MenuItem e Discount, o que permite análises detalhadas e suporte a operações diárias e estratégias comerciais. Por exemplo, cada linha detalhada (DetailLine) pode ser associada a um item ou a um desconto específico, permitindo rastrear quais produtos geram mais receita ou avaliar o impacto de promoções.

Além disso, a abordagem foi pensada para ser flexível e adaptável a mudanças futuras, como a introdução de novos tipos de dados ou alterações no esquema JSON original. Esse modelo garante a integridade dos dados com o uso de chaves primárias e estrangeiras e facilita a integração com sistemas de inteligência de negócios, atendendo tanto à análise granular quanto às necessidades de relatórios consolidados. Assim, é possível unir simplicidade operacional e robustez em um ambiente escalável.
